```python
#Import libraries
import requests
import bs4
import pandas as pd
```

```python
#Program

user_input = input("Enter 1 to find out current stock price or 2 to get AirNow data. If you want to run neither, enter no to quit.")

while user_input != "no":
    
    if user_input == "1":
        base_url = 'https://finance.yahoo.com/quote/{}?p={}'
        continue_on = True
        
        while continue_on:
            stock_ticker = input('Enter a stock ticker (symbol) or enter no to quit.')
            if stock_ticker == "no":
                continue_on = False
                user_input = input("Enter 1 to run Part One or 2 to run Part Two. If you want to run neither, enter no to quit.")
                
            formatted_url = base_url.format(stock_ticker,stock_ticker)
            result = requests.get(formatted_url)
            soup = bs4.BeautifulSoup(result.text,"lxml")
            price = soup.find_all('div',{'class': 'My(6px) Pos(r) smartphone_Mt(6px)'})[0].find('span').text
            print(stock_ticker)
            print(price)
            
    elif user_input == "2":
        continue_on = True
        
        while continue_on:
            zipCode = input('Enter the Zip Code or enter no to quit. ')
            if zipCode == "no":
                continue_on = False
                user_input = input("Enter 1 to run Part One or 2 to run Part Two. If you want to run neither, enter no to quit.")
            else:
                def main():
                    BaseURL="http://www.airnowapi.org/aq/forecast/zipCode/"
                    URLPost = {'API_KEY': '56D9CBE9-A49C-402D-BCCE-777A95E39DB3',
                            'zipCode': zipCode, 
                            'date': '2020-07-21',
                            'distance': '25',
                            'format': 'application/json'}
        
                    UseRequest(BaseURL, URLPost, zipCode)
            
                def UseRequest(BaseURL, URLPost, zipCode):
                    response=requests.get(BaseURL, URLPost)
                    jsontxt = response.json()
                    
                    for list in jsontxt:
                        State = list['StateCode']
                        Date = list['DateForecast']
                        City = list['ReportingArea']
                        AQIValue = list['AQI']
                        AQIValue_str = str(AQIValue)
                        AQIType = list['ParameterName']
                        Zip = zipCode.split()[-1]
            
                        State_list = [State]
                        Date_list = [Date]
                        City_list = [City]
                        AQIValue_list = [AQIValue]
                        AQIType_list = [AQIType]
                        Zip_list = [Zip]
        
                        df = pd.DataFrame([State_list,Date_list,City_list,AQIValue_list,AQIType_list,Zip_list],["State","Date","City","AQI Value","AQI Type","Zip Code"],['Results'])
                        print(df)
        
                        file = open("Part2_OUTFILE.txt", "a")
                        file.write("\n")
                        file.write("City: ")
                        file.write(City)
                        file.write("\n")
                        file.write("State: ")
                        file.write(State)
                        file.write("\n")
                        file.write("Zip Code: ")
                        file.write(Zip)
                        file.write("\n")
                        file.write("Date: ")
                        file.write(Date)
                        file.write("\n")
                        file.write("AQI Type: ")
                        file.write(AQIType)
                        file.write("\n")
                        file.write("AQI Value: ")
                        file.write(AQIValue_str)
                        file.write("\n")
                        file.close()
        
                main()
```
